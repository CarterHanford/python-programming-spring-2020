## Part 1 - geometry.py

```python
import arcpy
from arcpy import env

# set environment
arcpy.env.workspace = r'C:\Users\hanfordca\Documents\GitHub\python-programming-spring-2020\week 5 - geometries in Arcpy\data\Exercise08'

# set feature class and load cursor
fc = "rivers.shp"
cursor = arcpy.da.SearchCursor(fc, ["SHAPE@LENGTH"])
length = 0

# for loop - total length of all river segments
for row in cursor:
    length = length + row[0]

# step 2 with basic print function
# print(length)

# step 4 - print statement which turns output into a string
units = arcpy.Describe(fc).spatialReference.linearUnitName
print(str(length) + " " + units)
```

## Part 2 - points.py

```python
import arcpy
from arcpy import env

# set environment
arcpy.env.workspace = r'C:\Users\hanfordca\Documents\GitHub\python-programming-spring-2020\week 5 - geometries in Arcpy\data\Exercise08'

# set feature class and load cursor
fc = "dams.shp"
cursor = arcpy.da.SearchCursor(fc, ["SHAPE@XY"])

# for loop - prints x,y coordinates
for row in cursor:
    x, y = row[0]
    print("{0}, {1}".format(x,y))
```

## Part 3 - vertices.py

```python
import arcpy
from arcpy import env

# set environment
arcpy.env.workspace = r'C:\Users\hanfordca\Documents\GitHub\python-programming-spring-2020\week 5 - geometries in Arcpy\data\Exercise08'

# set feature class and load cursor
fc = "rivers.shp"
cursor = arcpy.da.SearchCursor(fc, ["OID@", "SHAPE@"])

for row in cursor:
    print("Feature {0}: ".format(row[0]))
    for point in row[1].getPart(0):
        print("{0}, {1}".format(point.X, point.Y))
```

## Part 4 - multipart.py

```python
import arcpy
from arcpy import env

# set environment
arcpy.env.workspace = r'C:\Users\hanfordca\Documents\GitHub\python-programming-spring-2020\week 5 - geometries in Arcpy\data\Exercise08'

# set feature class and load cursor
fc = "dams.shp"
cursor = arcpy.da.SearchCursor(fc, ["OID@", "SHAPE@"])

for row in cursor:
    if row[1].isMultipart:
        print("Feature {0} is multipart.".format(row[0]))
    else:
        print("Feature {0} is single part.".format(row[0]))
```

```python
import arcpy
from arcpy import env

# set environment
arcpy.env.workspace = r'C:\Users\hanfordca\Documents\GitHub\python-programming-spring-2020\week 5 - geometries in Arcpy\data\Exercise08'

# set feature class and load cursor
fc = "Hawaii.shp"
cursor = arcpy.da.SearchCursor(fc, ["OID@", "SHAPE@"])

for row in cursor:
    print("Feature {0}: ".format(row[0]))
    partnum = 0
    for part in row[1]:
        print("Part {0}: ".format(partnum))
        for point in part:
            print("{0}, {1}".format(point.X, point.Y))
        partnum += 1
```

## Part 5 - create.py

```python
import arcpy
import fileinput
import string
import os
from arcpy import env

arcpy.env.workspace = r'C:\Users\hanfordca\Documents\GitHub\python-programming-spring-2020\week 5 - geometries in Arcpy\data\Exercise08'
env.overwriteOutput = True
outpath = r'C:\Users\hanfordca\Documents\GitHub\python-programming-spring-2020\week 5 - geometries in Arcpy\data\Exercise08\Results'

newfc = "newpolyline.shp"
arcpy.CreateFeatureclass_management(outpath, newfc, "Polyline")

infile = r'C:\Users\hanfordca\Documents\GitHub\python-programming-spring-2020\week 5 - geometries in Arcpy\data\Exercise08\coordinates.txt'

polyline = r'C:\Users\hanfordca\Documents\GitHub\python-programming-spring-2020\week 5 - geometries in Arcpy\data\Exercise08\Results\newpolyline.shp'

cursor = arcpy.da.InsertCursor(polyline, ["SHAPE@"])
array = arcpy.Array()

for line in fileinput.input(infile):
    ID, X, Y = str.split(line," ")
    array.add(arcpy.Point(X, Y))

cursor.insertRow([arcpy.Polyline(array)])
fileinput.close()
del cursor
print("done.")
```
