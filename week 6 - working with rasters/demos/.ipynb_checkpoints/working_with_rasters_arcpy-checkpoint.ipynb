## [Describing Rasters](https://pro.arcgis.com/en/pro-app/arcpy/functions/raster-dataset-properties.htm)

```python
from arcpy.sa import *
arcpy.CheckOutExtension("Spatial")
```

```python
single_raster = "af_dem_30s.tif"
raster_info = arcpy.Describe(single_raster)
print(raster_info.format)
print(raster_info.spatialReference.name)
```

```python
print("Band Count:       %d" % raster_info.bandCount)
print("Compression Type: %s" % raster_info.compressionType)
print("Raster Format:    %s" % raster_info.format)
print("Permanent:        %s" % raster_info.permanent)
print("Sensor Type:      %s" % raster_info.sensorType)
```
